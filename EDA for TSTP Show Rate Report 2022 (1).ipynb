# QUESTIONS

Conditional Selection
#In performing exploratory data analysis, it is important to be able to select subsets of data to perform analysis or comparisons.

Which Month has the most sessions? 

```python
df1[df1.Month == df1.Month.max()]
```

```python
# what status is specific for the Month?

MonthMax = df1[df1.Month == df1.Month.max()]
MonthMax.Status
     
```

What Date in the Month?

```python
MonthMax.Date
```

Which month was the highest ?

```python
df1[df1['Month'] == df1['Month'].max()]
```
